```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
theta_arrDF = pd.read_csv('theta_arr.csv')
yDF = pd.read_csv('output.csv')
uDF = pd.read_csv('input.csv')

Ts = 0.185
final = 30
time = np.arange(0, final + Ts, Ts)
```

```python
def reference_signal(end_time=final, Ts_func=Ts, lower_set=0.2, upper_set=0.1, period=30):
    uc_func = []
    time = np.arange(0, end_time + Ts_func, Ts_func)
    for _t in time:
        rat = 2*np.pi/period
        if np.sin(rat*_t) >= 0:
            uc_func.append(upper_set)
        else:
            uc_func.append(lower_set)
    return np.array(uc_func, float)
uc_val = reference_signal()
# plt.plot(np.arange(0, 60 + Ts_val, Ts_val), uc)
# plt.show()
```

```python
filepath = "C:/Users/dboas065/University of Ottawa/ELG7113/experiment/code/Python_2022/pictures/"
theta = theta_arrDF[['a1', 'a2', 'b0', 'b1']].to_numpy()
for idx, row in enumerate(theta.T):
    plt.plot(time, row)
    plt.title('Parameter Adaptation vs. Time, t')
plt.xlabel('Time, t (sec)')
plt.ylabel('Parameter Values')
plt.legend(['$a_1$', '$a_2$', '$b_0$','$b_1$'])
plt.savefig(filepath+'adaptation_real.pdf', format='pdf')
plt.show()

y = yDF[['0']]

plt.plot(time, y)
plt.plot(time, reference_signal())
plt.title('System Response vs. Time, t')
plt.xlabel('Time, t (sec)')
plt.ylabel('Height from Bottom, h (m)')
plt.legend(['$y(t)$', '$u_c(t)$'])
plt.savefig(filepath+'output_real.pdf', format='pdf')
plt.show()

u = uDF[['0']]

plt.step(time[0:len(time)//1], u[0:len(time)//1])
plt.title('Input, $v_f(t) - v_{eq}$ vs. Time, t')
plt.xlabel('Time, t (sec)')
plt.ylabel('Air Speed, $v_f(t) - v_{eq}$ ($m/s^2$)')
plt.legend(['$u(t)$'])
plt.savefig(filepath+'input_real.pdf', format='pdf')
plt.show()
```
