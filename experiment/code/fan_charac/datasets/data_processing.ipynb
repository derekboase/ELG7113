```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset_1.csv')
df
```

```python
lst_duty = list(set(df['duty_cycle']))
lst_duty.sort()

lst_avg = [df['air_vel(m/s)'].loc[df['duty_cycle'] == lst_duty[i]].mean() for i in range(len(lst_duty))]
```

```python
df_avg = pd.DataFrame(np.array([lst_duty, lst_avg]).T, columns=['duty_cycle','air_vel_avg(m/s)'])
df_avg
```

```python
plt.rcParams['figure.figsize'] = [20, 10] # makes plots bigger

sns.set(font_scale = 2)
p = sns.lineplot(data=df_avg, x='duty_cycle', y='air_vel_avg(m/s)')
p.set_title('Average Air Velocity Vs. Duty Cycle')
```
