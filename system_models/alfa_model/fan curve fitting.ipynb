```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sympy as sp
```

```python
df = pd.read_csv('dataset_2__2022_03_07.csv')
df
```

```python
lst_duty = list(set(df['duty_cycle']))
lst_duty.sort()

lst_avg = [df['air_vel(m/s)'].loc[df['duty_cycle'] == lst_duty[i]].mean() for i in range(len(lst_duty))]
```

```python
df_avg = pd.DataFrame(np.array([lst_duty, lst_avg]).T, columns=['duty_cycle','air_vel_avg(m/s)'])
df_avg
```

```python
plt.rcParams['figure.figsize'] = [20, 10] # makes plots bigger

sns.set(font_scale = 2)
p = sns.lineplot(data=df_avg, x='duty_cycle', y='air_vel_avg(m/s)')
p.set_title('Average Air Velocity Vs. Duty Cycle')
```

```python
# fit a straight line to the economic data
from numpy import arange
from pandas import read_csv
from scipy.optimize import curve_fit
from matplotlib import pyplot
 
# define the true objective function
def objective(x, a, b):
    return a * x + b
 
# load the dataset
dataframe = df_avg
data = dataframe.values
# choose the input and output variables
x, y = dataframe["duty_cycle"], dataframe["air_vel_avg(m/s)"]
# curve fit
popt, _ = curve_fit(objective, x, y)
# summarize the parameter values
a, b = popt
# plot input vs output

pyplot.scatter(x, y)
# define a sequence of inputs between the smallest and largest known inputs
x_line = arange(min(x), max(x), 1)
# calculate the output for the range
y_line = objective(x_line, a, b)
# create a line plot for the mapping function
pyplot.plot(x_line, y_line, '--', color='red')
pyplot.show()
print('y = %.5f * x + %.5f' % (a, b))
```
